```python
import numpy as np
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
```

```python
train = pd.read_csv('/content/drive/MyDrive/ML/Kaggle/Diabetes/train.csv', index_col=0)
test = pd.read_csv('/content/drive/MyDrive/ML/Kaggle/Diabetes/test.csv', index_col=0)
test_ids = pd.read_csv('/content/drive/MyDrive/ML/Kaggle/Diabetes/test.csv')
test_ids = test_ids['id']

test.head(5)
```

```python
le = LabelEncoder()
columns = ['gender', 'ethnicity', 'education_level', 'income_level', 'smoking_status', 'employment_status']
for col in columns:
  train[col] = le.fit_transform(train[col])
  test[col] = le.fit_transform(test[col])

y = train['diagnosed_diabetes']
X = train.drop('diagnosed_diabetes', axis=1)

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, shuffle=False)
X_train.head()
```

```python
clf = GradientBoostingClassifier(n_estimators=50, learning_rate=0.01, max_depth=3).fit(X_train, y_train)
print(clf.score(X_val, y_val))

y_pred = clf.predict(test)
```

```python
submission_df = pd.DataFrame(data={"id": test_ids.values,
                           "diagnosed_diabetes": y_pred})
submission_df = submission_df.set_index('id')
print(submission_df)

submission_df.to_csv('/content/drive/MyDrive/ML/Kaggle/Diabetes/submission.csv')
```
